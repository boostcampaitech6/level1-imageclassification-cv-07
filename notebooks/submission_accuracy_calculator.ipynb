# 리더보드 예상 accuracy 계산기
### output.csv를 리더보드에 제출했을 때의 예상 accuracy를 계산해줍니다.

```python
import pandas as pd
```

```python
# control_group은 대조군입니다.
# 저희 리더보드의 1,2등 제출결과를 대조군으로하고, 예상 accuracy를 계산해줍니다.
# 36번째 제출과 37번째 제출결과를 다운받아서 대조군 path를 입력해주세요.
# exp는 실험군으로, 예상하고자하는 csv결과파일의 path를 입력해주세요.

control_group1_path = '/content/drive/MyDrive/aitech6/1_classification /output36.csv' # 36번째 대조군 주소
control_group2_path = '/content/drive/MyDrive/aitech6/1_classification /output37.csv' # 37번째 대조군 주소
exp_path = '/content/drive/MyDrive/aitech6/1_classification /mobilevitv2.csv'

df_cg1 = pd.read_csv(control_group1_path)
df_cg2 = pd.read_csv(control_group2_path)
df_exp = pd.read_csv(exp_path)

df_exp
```

```python
# 대조군과 실험군의 ans를 비교합니다.
df_cg1['ans']==df_exp['ans']
df_cg2['ans']==df_exp['ans']
```

```python
# n1,n2는 각각의 대조군과 실험군을 비교해서 구한 ans 일치 개수입니다.
n1 = df_exp[df_cg1['ans']==df_exp['ans']].ans.count()
n2 = df_exp[df_cg2['ans']==df_exp['ans']].ans.count()
total = df_exp.ans.count()
print('기준1과 일치하는 데이터 개수:',n1)
print('기준2과 일치하는 데이터 개수:',n2)
print('데이터 전체 개수:',total)
```

```python
# 대조군과 일치하는 데이터 개수를 데이터(eval) 전체 개수로 나누어줍니다.
acc1 = n1/total
acc2 = n2/total
corr_cons1 = 1.1 # correction_constant
corr_cons2 = 1.25
print(f'기준1과 일치 비율: { acc1*100:.3f}%')
print(f'기준1과 예상 성적 : {0.795 * acc1 * corr_cons1:.3f}')
print(f'기준2과 일치 비율: { acc2*100:.3f}%')
print(f'기준2과 예상 성적 : {0.795 * acc2 * corr_cons2:.3f}')
print()

print(f'최종 예상 성적 : {0.795 * (acc1 * corr_cons1 + acc2 * corr_cons2) / 2 :.3f}')
```
